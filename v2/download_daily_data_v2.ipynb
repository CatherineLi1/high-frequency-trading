```python
from data_layer_v2 import IBKRDataClient
import pandas as pd

client = IBKRDataClient()

symbols = ["KO", "PEP", "XOM", "CVX", "JPM", "BAC", "HD", "LOW"]

contracts = client.qualify(symbols)

def download_daily(symbol, contract):

    print(f"Downloading {symbol}...")

    bars = client.ib.reqHistoricalData(
        contract,
        endDateTime="",
        durationStr="10 Y",
        barSizeSetting="1 day",
        whatToShow="TRADES",
        useRTH=True,
        formatDate=1
    )

    df = pd.DataFrame(bars)
    df = df.rename(columns={
        "date": "date",
        "open": "open",
        "high": "high",
        "low": "low",
        "close": "close",
        "volume": "volume"
    })

    df = df.set_index("date")
    df.index = pd.to_datetime(df.index)

    df.to_csv(f"{symbol}.csv")

    print(f"{symbol} saved.\n")

for symbol in symbols:
    download_daily(symbol, contracts[symbol])

client.disconnect()

print("All industry daily data downloaded.")
```
